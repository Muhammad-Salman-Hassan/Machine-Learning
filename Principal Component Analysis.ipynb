```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading Breast cancer from sklearn,sklearn provide built in data sets on it to learn 
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS.fit(df)
scaled_data=SS.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3,random_state=0)
pca_scaled=pca.fit_transform(scaled_data)
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cancer['target'])
```

```python
plt.scatter(pca_scaled[:,1],pca_scaled[:,2],c=cancer['target'])
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,2],c=cancer['target'])
```

```python
plt.figure(figsize=(10,15))
ax=plt.axes(projection='3d')
ax.scatter3D(pca_scaled[:,0],pca_scaled[:,1],pca_scaled[:,2],c=cancer['target'],alpha=0.5)
```
